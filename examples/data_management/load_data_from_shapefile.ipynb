# Load data from a Shapefile

This example illustrates how to load data from a shapefile using GeoPandas.
A shapefile is a complex format, compared to CSV or GeoJSON, so [GeoPandas offers more options](https://geopandas.org/io.html#reading-spatial-data) for reading it.

```python
from geopandas import read_file

gdf = read_file('https://libs.cartocdn.com/cartoframes/files/sustainable_palm_oil_production_mills.zip')
gdf.head()
```

```python
from cartoframes.viz import Layer

Layer(gdf)
```
