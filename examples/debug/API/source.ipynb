```python
from cartoframes.auth import set_default_context
from cartoframes.viz import Map, Layer, Source

set_default_context(
    base_url='https://cartovl.carto.com/', 
    api_key='default_public'
)
```

```python
# Source: Table name
map = Map(
    Layer(
        Source('populated_places')
    )
)

# Source: Table name + sugar
Map(Layer('populated_places'))
```

```python
# Source: SQL query
map = Map(
    Layer(
        Source('SELECT * FROM populated_places LIMIT 10')
    )
)

# Source: SQL query + sugar
Map(Layer('SELECT * FROM populated_places LIMIT 10'))
```

```python
# Source: GeoJSON
map = Map(
    Layer(
        Source('points.geojson')
    )
)

# Source: GeoJSON + sugar
Map(Layer('points.geojson'))
```

```python
# Source: DataFrame

import pandas as pd

data = {'latitude': [0, 10, 20, 30], 'longitude': [0, 10, 20, 30]}
df = pd.DataFrame.from_dict(data)

Map(Layer(df, 'width: 30'))
```

```python
# Source: DataFrame (from CARTO)

from cartoframes.data import Dataset

ds = Dataset.from_table('populated_places')

df = ds.download(limit=10)

Map(Layer(df, 'width: 10'))
```

```python
# Source: Dataset

from cartoframes.data import Dataset

ds = Dataset.from_table('populated_places')

Map(Layer(ds))
```

```python
# Source: Dataset

import pandas as pd

data = {'latitude': [0, 10, 20, 30], 'longitude': [0, 10, 20, 30]}
df = pd.DataFrame.from_dict(data)

ds = Dataset.from_dataframe(df)

Map(Layer(ds, 'width: 30'))
```
