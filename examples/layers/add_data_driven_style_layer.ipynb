# Add data-driven style to a Layer

In this example, a data-driven style is added to the Layer with the `color` property.

For more information, run `help(Layer)`.

```python
from cartoframes.auth import set_default_credentials

set_default_credentials('cartoframes')
```

```python
from cartoframes import CartoDataFrame
from cartoframes.viz import Layer

cdf = CartoDataFrame.from_carto("SELECT * FROM global_power_plants WHERE country IN ('Brazil')")

Layer(
    cdf,
    '''
    color: ramp(buckets($fuel_type,['Hydro','Oil','Biomass']),[turquoise,orange,violet])
    width: 5
    strokeColor: rgb(255,255,255)
    strokeWidth: 1
    '''
)
```
